```python
import os
import pandas as pd
import geopandas as gpd
import numpy as np
from netCDF4 import Dataset
from core.abi import remap
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
```

```python
df = gpd.read_file('south-america/SouthAmerica.shp')
df.geometry.bounds
```

```python
path = 'data/noaa-goes16/ABI-L2-CMIPF/2019/329/13'
files = [os.path.join(path, file) for file in os.listdir(path)]
files = list(filter(lambda file: file.endswith('.nc'), files))
files = {file[60:63]: file for file in sorted(files)}
bboxsa = [-85, -33, -56, 15]
```

```python
rch1 = remap(path=files['C13'], extent=bboxsa, resolution=2, driver='netCDF4')
```

```python
plt.figure(figsize=(10, 17))

gch1 = rch1.ReadAsArray()
lons = np.linspace(bboxsa[0], bboxsa[2], gch1.shape[1])
lats = np.linspace(bboxsa[1], bboxsa[3], gch1.shape[0])
# lons, lats = np.meshgrid(lons, lats)
m = Basemap(projection='merc', llcrnrlat=bboxsa[1], urcrnrlat=bboxsa[3],\
            llcrnrlon=bboxsa[0], urcrnrlon=bboxsa[2], resolution='c')
# m = Basemap(projection='merc',llcrnrlat=-80,urcrnrlat=80,\
#             llcrnrlon=-180,urcrnrlon=180,lat_ts=20,resolution='c')
m.drawcoastlines()
# plt.pcolormesh(gch1, lons, lats, zorder=99)
```
